# Change Stomatal Resistance Parameterization from Ball-berry to Jarvis

```python
object_dir = "/home/jovyan/work/sciunit/MyAnalysis/cde-package/cde-root/home/choi/sciunit/Bakinam/"
```

```python
simulation_object = object_dir+"simulation_object.py"
```

```python
!cat {simulation_object}
```

- Change Stomatal resistance parameterization from "Ball-berry" to "Jarvis"

```python
with open(simulation_object, "r") as f:
    read_data = f.read()
    line = read_data.replace('BallBerry', "Jarvis")
    f.close()
with open(simulation_object, "w") as f:
    f.write(line)
    f.close()
```

```python
!cat {simulation_object}
```
